```python
from shapely.geometry import Point

# An inspector's location (somewhere in Chicago)
inspector = Point(-87.7650, 41.9658)  # lon, lat — note the order!

# An order location (let's say somewhere in Illinois)
order = Point(-88.1000, 41.8000)

# Euclidean "distance" — in degrees, not miles (we'll fix this soon)
print(f"Raw distance: {inspector.distance(order):.4f} degrees")
```

```python
from shapely.geometry import Point

# ProxyPics HQ — note lon FIRST, then lat
hq = Point(-87.7650, 41.9658)

# A pretend order location in downstate Illinois
order = Point(-89.0000, 40.5000)

# Raw Shapely distance (in degrees — not miles yet)
raw_dist = hq.distance(order)
print(f"Shapely distance (degrees): {raw_dist:.4f}")

# Quick sanity check: print the coordinates back out
print(f"HQ coordinates: {hq.x}, {hq.y}")
print(f"Order coordinates: {order.x}, {order.y}")
```

```python
from shapely.geometry import Point

# Inspector somewhere in rural Illinois
inspector = Point(-89.0000, 40.5000)

# Draw a "circle" around them — radius in degrees
# ~0.145 degrees ≈ 10 miles at this latitude (we'll make this precise shortly)
coverage_zone = inspector.buffer(0.145)

# What did buffer() return?
print(type(coverage_zone))
print(f"Is it a polygon? {coverage_zone.geom_type}")
print(f"Number of boundary points: {len(coverage_zone.exterior.coords)}")

# Now the useful question: is an order inside this inspector's zone?
order_inside  = Point(-88.9000, 40.4500)   # close — should be IN
order_outside = Point(-88.0000, 40.5000)   # far — should be OUT

print(f"\nNear order inside zone: {coverage_zone.contains(order_inside)}")
print(f"Far order inside zone:  {coverage_zone.contains(order_outside)}")
```

```python
import math

def degrees_from_miles(miles, latitude):
    """Convert a mile radius to approximate degrees for Shapely buffer()"""
    lat_rad = math.radians(latitude)
    # Average of lat and lon degree lengths at this latitude
    miles_per_degree = 69.0 * (1 + math.cos(lat_rad)) / 2
    return miles / miles_per_degree

# Test it for Chicago
radius_10mi  = degrees_from_miles(10, 41.9658)
radius_20mi  = degrees_from_miles(20, 41.9658)

print(f"10-mile buffer radius: {radius_10mi:.4f} degrees")
print(f"20-mile buffer radius: {radius_20mi:.4f} degrees")

# Now rebuild the inspector zone properly
inspector = Point(-89.0000, 40.5000)
zone_10mi = inspector.buffer(radius_10mi)
zone_20mi = inspector.buffer(radius_20mi)

print(f"\n10-mile zone area: {zone_10mi.area:.6f} sq degrees")
print(f"20-mile zone area: {zone_20mi.area:.6f} sq degrees")
```
